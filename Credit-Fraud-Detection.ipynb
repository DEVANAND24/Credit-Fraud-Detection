```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a Pandas Dataframe
creditcard = pd.read_csv('/content/creditcard.csv')
```

```python
# first 5 row of dataset
creditcard.head()
```

```python
#last 5 row of dataset
creditcard.tail()
```

```python
#dataset information
creditcard.info()
```

```python
#checking number of missing value in each column
creditcard.isnull().sum()
```

```python
 #dropping the missing values
creditcard = creditcard.dropna(how='any')
```

```python
creditcard.isnull().sum()
```

```python
#distribution of legit transaction and fraudulent transaction
creditcard['Class'].value_counts()
```

```python
#distribution of legit transaction and fraudulent transaction
creditcard['Class'].value_counts()
```

```python
#seprating the data for analysis
legit = creditcard[creditcard.Class == 0]
fraud = creditcard[creditcard.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#comparing the value for both transaction
creditcard.groupby('Class').mean()
```

```python
#build a sample dataframe that is equivalent of fraudulent dataset
legit_sample=legit.sample(n=104)
```

```python
#concataining the legit and fraud dataset
new_dataset  = pd.concat([legit_sample,fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#splitting data in features and target
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#splitting the data for training and testing
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape, X_test.shape)
```

Model training


```python
#logistic Regression
model = LogisticRegression()
```

```python
#training the logistic Regression model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score


```python
#Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train.prediction, Y_train)
```

```python
print('Accuracy on training data:', training_data_accuracy)
```

```python
#Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data:' , test_data_accuracy)
```
